```python
import plotly.express as px
from sklearn.decomposition import PCA

df_pca_data = pd.read_csv('pima-indians-diabetes.csv'  )
features = ["Number_pregnant","Plasma_glucose","Blood_pressure","Triceps_thickness","Serum_insulin","BMI","Diabetes_function","Age","Class_variable"]

pca = PCA()
components = pca.fit_transform(df_pca_data[features])
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(4),
    color=df_pca_data["Serum_insulin"]
)
fig.update_traces(diagonal_visible=False)
fig.show()
```

```python
import plotly.express as px
from sklearn.decomposition import PCA

df_pca_data = pd.read_csv('pima-indians-diabetes.csv'  )
features =df_pca_data[["Number_pregnant","Plasma_glucose","Blood_pressure","Triceps_thickness","Serum_insulin","BMI","Diabetes_function","Age","Class_variable"]]

pca = PCA(n_components=3)
components = pca.fit_transform(features)

total_var = pca.explained_variance_ratio_.sum() * 100

fig = px.scatter_3d(
    components, x=0, y=1, z=2, color=df_pca_data['Serum_insulin'],
    title=f'Total Explained Variance: {total_var:.2f}%',
    labels={'0': 'PC 1', '1': 'PC 2', '2': 'PC 3'}
)
fig.show()
```
